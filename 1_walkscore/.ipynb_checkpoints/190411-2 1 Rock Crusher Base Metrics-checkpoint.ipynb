# Crusher Actual Metrics

Compute the actual opportunity using crusher metallurgy and production data

## Load Factors

### Average Number of Loads Per Shift: 

```python
%%cypher http://neo4j:password@localhost:7474/db/data
Match(s:Shift)-[]-(p:ProductionEntry)-[]-(l:Location) where l.Name='X-Pit Crusher'  
with 
s.Name as Shift,
s.ShiftDate as Date,
l,
sum(p.Loads) as Loads
return 

l.Name as Location,
Shift,
avg(Loads)
```

### Average Number of Loads Per Day: 

```python
%%cypher http://neo4j:password@localhost:7474/db/data
Match(s:Shift)-[]-(p:ProductionEntry)-[]-(l:Location) where l.Name='X-Pit Crusher' and s.ShiftDate.year=2018
with 
s.ShiftDate as Date,
sum (p.Loads) as loads 
return 
avg(loads) as averageLoadsPerDay
```

### Calculate Actual Average Tons Per Hour

Average Loads Per Day/24 hrs * 150 Tonnes per truck

```python
%%cypher http://neo4j:password@localhost:7474/db/data
Match(s:Shift)-[]-(p:ProductionEntry)-[]-(l:Location) where l.Name='X-Pit Crusher' and s.ShiftDate.year=2018
with 
s.ShiftDate as Date,
sum (p.Loads) as loads 
return 
avg(loads)/24*150 as ActualAverageTonnesPerHour
```

## Metallurgical Data for Actual Grades

Using the Metallurgy data set, we can calculate the actual average grades of Silver and Gold for the avaialable production data. The following result set also computes standard deviation for months of the year for review: 

```python
%%cypher http://neo4j:password@localhost:7474/db/data
MATCH (m:MetallurgyMeasurement) 
return 
m.ShiftDate.year as year,
m.ShiftDate.month as month,
avg(m.GoldGrade) as averageGoldGrade, 
avg(m.SilverGrade) as averageSilverGrade,
stDev(m.GoldGrade) as standardDeviationGoldGrade,
stDev(m.SilverGrade) as standardDeviationSilverGrade
```

Using these data, we can compute the actual average Silver and Gold grades: 

```python
%%cypher http://neo4j:password@localhost:7474/db/data
MATCH (m:MetallurgyMeasurement)  
return 
avg(m.GoldGrade) as averageGoldGrade, 
avg(m.SilverGrade) as averageSilverGrade,
stDev(m.GoldGrade) as standardDeviationGoldGrade,
stDev(m.SilverGrade) as standardDeviationSilverGrade
```

These match our estimates very closely, but while the standard deviation calculation shows that while the gold grades are consistent through the year, the silver grades have a much higher standard deviation and therefore are more variable. This is likley due to source material grade variability.
